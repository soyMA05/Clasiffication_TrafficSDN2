```python
#librerias 
import pandas as pd
import numpy as np
import pathlib as pl
from os import path
from libmac import listar_archivos, etiquetar_datos, concatenar_datos
```

```python
#diccionario de acceso a archivos de extraccion de flujos
dir_trafico = pl.Path("D:/ML-Tesis-Attacks/PcapToCsv/ataqueovs/")
"""crear diccionario para el acceso de imagenes"""
dict_ataque = {
    'BOTNET': list(dir_trafico.glob('Botnet/*.csv')),
    'BFA': list(dir_trafico.glob('brute-force-attack/*.csv')),  
    'DDoS': list(dir_trafico.glob('DDoS/*.csv')),  
    'Dos': list(dir_trafico.glob('dos/**/*.csv')),  
    'Probe': list(dir_trafico.glob('Probe/*.csv')),  
    'Web-Attack': list(dir_trafico.glob('web_attack/*.csv'))  
}
```

```python
#mostrar archivos
listar_archivos(dict_ataque)
```

```python
#etiquetar tipo de trafico segun la clase especificada
def etiquetar_datos(_dictdatos:dict, caracteristica:str):
    #dataframes = [] #lista para crear dataframes
    for clase, carpeta in _dictdatos.items():
        for archivo in range(len(carpeta)):
            archivo_exdata = str(_dictdatos[clase][archivo])
            print('En proceso...etiquetando datos de %s',archivo_exdata)
            df = pd.read_csv(archivo_exdata)
            df[caracteristica]=clase
            df.to_json(archivo_exdata.replace(".csv",".json"),orient='records',lines=True)
```
